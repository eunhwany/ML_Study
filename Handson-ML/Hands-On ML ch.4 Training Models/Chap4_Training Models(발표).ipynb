```python
import numpy as np
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)
```

```python
plt.plot(X, y, "b.")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.axis([0, 2, 0, 15])
plt.show()
```

```python
# 정규방정식을 이용하여 계산.
X_b = np.c_[np.ones((100,1)),X] # X_0은 항상 1
# linalg : numpy의 Linear algebra > 역행렬 사용위해 
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
```

```python
# 예측된 linear 직선 예측
plt.plot(X_new, y_predict, "r-", linewidth=2, label="Predictions")
plt.plot(X, y, "b.")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.legend(loc="upper left", fontsize=14)
plt.axis([0, 2, 0, 15])
plt.show()
```

### Computational Complexity
- 정규방정식은 X(n개의 feature 개수를 가진 행렬)에 대해 X^T * X(n * n 형)의 역행렬을 계산하는데 이 행렬의 역행렬을 계산시 O(n^2.4)에서 O(n^3)정도의 계산복잡도를 가진다.(feature의 개수가 2배로 늘어나면 > 2^2.4 = 5.3, 2^3=8을 곱함)
- 긍정적인 측면 : 정규방정식의 학습 set의 변수에 대해 선형적이라 효율적으로 변수를 다룰 수 있고 예측이 빠르다.
- 부정적인 측면 :  정규방정식인 경우 feature의 개수가 많을때(10만개정도), 계산이 느리다. feature의 개수가 많은 모델에 대해서는 적합하지 않을 수 있다.
> 이를 해결하기 위한 것이 'Gredient Descent(경사하강법)'

### 2) Linear Regression using gredient descent
- cost function(비용함수)를 최소화 하기 위해 매개변수를 반복적으로 조정하는 것.
- 광범위한 최적의 해(optimal solutions)를 찾을 수 있는 일반적인 최적화 알고리즘.
- MSE(Mean Square Error)를 최소화하는 $theta$를 정하는것!
![](./imag/gd1.png)

- 그림의 중요 매개변수 : learing rate(학습속도)의 크기!

- learning rate가 작을 경우

- learning rate가 클 경우 : 발산되어 최적솔루션에 도달을 못함.
![](./imag/maxgd.png)

- 불규칙한 지형의 경우
![](./imag/irrgd.png)

- plateau(고원) >> 최소한의 값으로 수렴하는 것을 어렵게 만듬.
- 전역최소값(global minimum)이 아닌 지역최소값(local mininmum)으로 수렴될 수도 있음.

- 선형회귀 모형의 MSE 비용함수는 : 볼록(convex)한 함수
    - 하나의 전역최소값.
    - 경사가 변하지 않는 연속함수.
![](./imag/convex.png)

- 왼쪽 : 동일한 scale, 오른쪽 : scale이 다른 경우 길어질 수 있음
- feature scale이 중요함(scikit-learn의 Standard Scaler) 그렇지 않으면 시간이 오래 걸릴 수 있음.

#### Batch Gredient Descent

- 경사 하강 구현시, 각각의 모델 매개변수 $theta_j$에 대한 비용함수를 각각 계산
- $theta_j$를 조금 변경했을 때 비용함수는 어떻게 달라지는지 계산 > 편미분(partial derivative) 
![](./imag/batchgd1.png)
![](./imag/batchgd2.png)

- 왜? Batch Gredient Descent(일괄 경사 하강) 이라 불리는가?
    - the whole batch of training data at every step : 각 경사 하강단계에서 학습 set X의 전체에 대해 계산을 해야되기 때문.
    - As a result it is terribly slow on very large training sets: 결과적으로 매우 큰 학습 set에 대해서는 엄청 느릴 수밖에 없다.

- 다음 단계 > 내리막으로 가기 위한 식
![](./imag/batchgd3.png)

```python
learning_rate = 0.1 # 학습속도
n_iterations = 1000
m = 100 # instance의 개수
theta = np.random.randn(2,1)

for iteration in range(n_iterations):
    gradients = 2/m * X_b.T.dot(X_b.dot(theta) - y)
    theta = theta - learning_rate * gradients
theta # 정규방정식에서 구한값과 같다.
```

```python
X_new_b.dot(theta)
```

```python
# 학습속도 변경시
theta_path_bgd = []

def plot_gradient_descent(theta, learning_rate, theta_path=None):
    m = len(X_b)
    plt.plot(X, y, "g.")
    n_iterations = 1000
    for iteration in range(n_iterations):
        if iteration < 10:
            y_predict = X_new_b.dot(theta)
            style = "b-" if iteration > 0 else "r--"
            plt.plot(X_new, y_predict, style)
        gradients = 2/m * X_b.T.dot(X_b.dot(theta) - y)
        theta = theta - learning_rate * gradients
        if theta_path is not None:
            theta_path.append(theta)
    plt.xlabel("$x_1$", fontsize=18)
    plt.axis([0, 2, 0, 15])
    plt.title(r"$\eta = {}$".format(learning_rate), fontsize=16)

np.random.seed(42)
theta = np.random.randn(2,1)  # random initialization

plt.figure(figsize=(10,4))
plt.subplot(131); plot_gradient_descent(theta, learning_rate=0.02)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.subplot(132); plot_gradient_descent(theta, learning_rate=0.1, theta_path=theta_path_bgd)
plt.subplot(133); plot_gradient_descent(theta, learning_rate=0.5)
plt.show()
```

#### Stochastic(확률론적인) Gredient Descent
- 모든 단계에서 학습 set을 random instance를 선택하고(stochastic: 확률론적), 해당 instance만 기울기를 계산 > 빠르다, 각 반보가다 하나의 instance만 메모리에서 작동하기 때문에 대량의 학습 set 구동 가능.
- Batch Gredient Descent보다 덜 규칙적.
- 비용함수가 최소에 도달할 때까지 평균적으로만 감소함 >> 시간이 지남에 따라 최소값에 매우 근접하지만 일정한 점에서 끝나는 것이 아니라 주변을 맴돌게 될 것임 > 최적은 아닐 수 있다.
![](./imag/stobatchgd.png)

- 학습속도를 조절 (초기: 빠르게(지역최소값에서 벗어날 수 있음) > 점차적으로 낮춤(전역 최소값에 정착)) >> 학습 스케쥴(Learning Schedule)

```python
theta_path_sgd = []
m = len(X_b)
np.random.seed(42)
```

```python
n_epochs = 50
t0, t1 = 5, 50  # learning schedule hyperparameters

def learning_schedule(t):
    return t0 / (t + t1)

theta = np.random.randn(2,1)  # random initialization

for epoch in range(n_epochs):
    for i in range(m):
        if epoch == 0 and i < 20:                    # not shown in the book
            y_predict = X_new_b.dot(theta)           # not shown
            style = "b-" if i > 0 else "r--"         # not shown
            plt.plot(X_new, y_predict, style)        # not shown
        random_index = np.random.randint(m)
        xi = X_b[random_index:random_index+1]
        yi = y[random_index:random_index+1]
        gradients = 2 * xi.T.dot(xi.dot(theta) - yi)
        eta = learning_schedule(epoch * m + i)
        theta = theta - eta * gradients
        theta_path_sgd.append(theta)                 # not shown

plt.plot(X, y, "b.")                                 # not shown
plt.xlabel("$x_1$", fontsize=18)                     # not shown
plt.ylabel("$y$", rotation=0, fontsize=18)           # not shown
plt.axis([0, 2, 0, 15])                              # not shown
plt.show()                                           # not shown
```

#### Mini-batch Gradient Descent
- 각 단계에서 전체에 대한 (batch) 또는 하나의 instance(stochastic)를 기반으로 하는 것 대신 mini-batch라고 하는 instance의 작은 무작위 set에서 기울기를 계산.
- stochastic보다 덜 불규칙, 최종적으로 최소값에 더 가깝게 끝날 수 있음.

![](./imag/minibatch.png)

```python
theta_path_mgd = []

n_iterations = 50
minibatch_size = 20

np.random.seed(42)
theta = np.random.randn(2,1)  # random initialization

t0, t1 = 10, 1000
def learning_schedule(t):
    return t0 / (t + t1)

t = 0
for epoch in range(n_iterations):
    shuffled_indices = np.random.permutation(m)
    X_b_shuffled = X_b[shuffled_indices]
    y_shuffled = y[shuffled_indices]
    for i in range(0, m, minibatch_size):
        t += 1
        xi = X_b_shuffled[i:i+minibatch_size]
        yi = y_shuffled[i:i+minibatch_size]
        gradients = 2 * xi.T.dot(xi.dot(theta) - yi)
        eta = learning_schedule(t)
        theta = theta - eta * gradients
        theta_path_mgd.append(theta)
```

```python
heta_path_bgd = np.array(theta_path_bgd)
theta_path_sgd = np.array(theta_path_sgd)
theta_path_mgd = np.array(theta_path_mgd)
```

```python
plt.figure(figsize=(7,4))
plt.plot(theta_path_sgd[:, 0], theta_path_sgd[:, 1], "r-s", linewidth=1, label="Stochastic")
plt.plot(theta_path_mgd[:, 0], theta_path_mgd[:, 1], "g-+", linewidth=2, label="Mini-batch")
plt.plot(theta_path_bgd[:, 0], theta_path_bgd[:, 1], "b-o", linewidth=3, label="Batch")
plt.legend(loc="upper left", fontsize=16)
plt.xlabel(r"$\theta_0$", fontsize=20)
plt.ylabel(r"$\theta_1$   ", fontsize=20, rotation=0)
plt.axis([2.5, 4.5, 2.3, 3.9])
plt.show()
```

### 4) Learning Curves

![](./imag/lcv.png)
- 선형회귀모델 : underfitting
- 300차 모델 : overfitting 
    - 일반화 된 모델 : 2차원 모델.

- Cross-Validation(교차검증) : 모델의 일반화 성능 추정방법(Chap2 참조)
- 학습 set의 서로 다른 크기의 subset을 여러번 훈련시켜야함 > learing curve를 살펴봄

```python
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

def plot_learning_curves(model, X, y):
    X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=10)
    train_errors, val_errors = [], []
    for m in range(1, len(X_train)):
        model.fit(X_train[:m], y_train[:m])
        y_train_predict = model.predict(X_train[:m])
        y_val_predict = model.predict(X_val)
        train_errors.append(mean_squared_error(y_train_predict, y_train[:m]))
        val_errors.append(mean_squared_error(y_val_predict, y_val))

    plt.plot(np.sqrt(train_errors), "r-+", linewidth=2, label="train")
    plt.plot(np.sqrt(val_errors), "b-", linewidth=3, label="val")
    plt.legend(loc="upper right", fontsize=14)   # not shown in the book
    plt.xlabel("Training set size", fontsize=14) # not shown
    plt.ylabel("RMSE", fontsize=14)              # not shown
```

- underfitting : 모델을 복잡하게 바꾸거나 좋은 feature를 추가해주는 방법.

```python
# 10차 다항식 모델의 학습곡선
from sklearn.pipeline import Pipeline

polynomial_regression = Pipeline([
        ("poly_features", PolynomialFeatures(degree=10, include_bias=False)),
        ("lin_reg", LinearRegression()),
    ])

plot_learning_curves(polynomial_regression, X, y)
plt.axis([0, 80, 0, 3])           # not shown
plt.show()                        # not shown
```

### 5) Regularized Linear Models
- overfitting을 줄이는 방법의 하나로써 모델을 정규화(Regularize)하는 방법이 있음.
- Regularization: 모든 특징을 사용하나 특징 $θ_i$에 대한 parameter를 줄인다
- 선형 모델의 정규화는 모델의 가중치를 제한하여 수행.
- Ridge / Lasso Regression, Elastic net

#### Ridge Regression
- 비용함수에 아래의 식의 오른쪽 부분과 같은 정규화항을 적용
- 학습 알고리즘이 데이터에 적합 & 모델 가중치를 가능한 작게 유지하게 됨.
![](./imag/ridge.png)
- α=0 : 단순선형회귀
- α를 크게 하면 모든 가중치는 0에 가까워져서 > 데이터의 평균을 지나는 평평한 직선.

- feature 가중치 벡터(Θ1~Θn)을 vector W로 정의 > 정규화항은 $1/2*(||w||_2)^2$
- 경사하강법은 MSE 기울기 벡터에 αW만 추가.
- 주의점 : 입력 feature의 scale에 민감하기 때분에 StandardScaler같은 데이터의 scale하는 것이 중요 > 대부분의 정규화된 모델에 해당.

```python
from sklearn.linear_model import Ridge
from sklearn.preprocessing import StandardScaler

np.random.seed(42)
m = 20
X = 3 * np.random.rand(m, 1)
y = 1 + 0.5 * X + np.random.randn(m, 1) / 1.5
X_new = np.linspace(0, 3, 100).reshape(100, 1)

def plot_model(model_class, polynomial, alphas, **model_kargs):
    for alpha, style in zip(alphas, ("b-", "g--", "r:")):
        model = model_class(alpha, **model_kargs) if alpha > 0 else LinearRegression()
        if polynomial:
            model = Pipeline([
                    ("poly_features", PolynomialFeatures(degree=10, include_bias=False)),
                    ("std_scaler", StandardScaler()),
                    ("regul_reg", model),
                ])
        model.fit(X, y)
        y_new_regul = model.predict(X_new)
        lw = 2 if alpha > 0 else 1
        plt.plot(X_new, y_new_regul, style, linewidth=lw, label=r"$\alpha = {}$".format(alpha))
    plt.plot(X, y, "b.", linewidth=3)
    plt.legend(loc="upper left", fontsize=15)
    plt.xlabel("$x_1$", fontsize=18)
    plt.axis([0, 3, 0, 4])

plt.figure(figsize=(8,4))
plt.subplot(121)
plot_model(Ridge, polynomial=False, alphas=(0, 10, 100), random_state=42)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.subplot(122)
plot_model(Ridge, polynomial=True, alphas=(0, 10**-5, 1), random_state=42)
plt.show()
```

- α를 증가시 예측이 더욱 합리적해 지는 것을 확인.

```python
t1a, t1b, t2a, t2b = -1, 3, -1.5, 1.5

# ignoring bias term
t1s = np.linspace(t1a, t1b, 500)
t2s = np.linspace(t2a, t2b, 500)
t1, t2 = np.meshgrid(t1s, t2s)
T = np.c_[t1.ravel(), t2.ravel()]
Xr = np.array([[-1, 1], [-0.3, -1], [1, 0.1]])
yr = 2 * Xr[:, :1] + 0.5 * Xr[:, 1:]

J = (1/len(Xr) * np.sum((T.dot(Xr.T) - yr.T)**2, axis=1)).reshape(t1.shape)

N1 = np.linalg.norm(T, ord=1, axis=1).reshape(t1.shape)
N2 = np.linalg.norm(T, ord=2, axis=1).reshape(t1.shape)

t_min_idx = np.unravel_index(np.argmin(J), J.shape)
t1_min, t2_min = t1[t_min_idx], t2[t_min_idx]

t_init = np.array([[0.25], [-1]])
```

```python
def bgd_path(theta, X, y, l1, l2, core = 1, eta = 0.1, n_iterations = 50):
    path = [theta]
    for iteration in range(n_iterations):
        gradients = core * 2/len(X) * X.T.dot(X.dot(theta) - y) + l1 * np.sign(theta) + 2 * l2 * theta

        theta = theta - eta * gradients
        path.append(theta)
    return np.array(path)

plt.figure(figsize=(12, 8))
for i, N, l1, l2, title in ((0, N1, 0.5, 0, "Lasso"), (1, N2, 0,  0.1, "Ridge")):
    JR = J + l1 * N1 + l2 * N2**2
    
    tr_min_idx = np.unravel_index(np.argmin(JR), JR.shape)
    t1r_min, t2r_min = t1[tr_min_idx], t2[tr_min_idx]

    levelsJ=(np.exp(np.linspace(0, 1, 20)) - 1) * (np.max(J) - np.min(J)) + np.min(J)
    levelsJR=(np.exp(np.linspace(0, 1, 20)) - 1) * (np.max(JR) - np.min(JR)) + np.min(JR)
    levelsN=np.linspace(0, np.max(N), 10)
    
    path_J = bgd_path(t_init, Xr, yr, l1=0, l2=0)
    path_JR = bgd_path(t_init, Xr, yr, l1, l2)
    path_N = bgd_path(t_init, Xr, yr, np.sign(l1)/3, np.sign(l2), core=0)

    plt.subplot(221 + i * 2)
    plt.grid(True)
    plt.axhline(y=0, color='k')
    plt.axvline(x=0, color='k')
    plt.contourf(t1, t2, J, levels=levelsJ, alpha=0.9)
    plt.contour(t1, t2, N, levels=levelsN)
    plt.plot(path_J[:, 0], path_J[:, 1], "w-o")
    plt.plot(path_N[:, 0], path_N[:, 1], "y-^")
    plt.plot(t1_min, t2_min, "rs")
    plt.title(r"$\ell_{}$ penalty".format(i + 1), fontsize=16)
    plt.axis([t1a, t1b, t2a, t2b])

    plt.subplot(222 + i * 2)
    plt.grid(True)
    plt.axhline(y=0, color='k')
    plt.axvline(x=0, color='k')
    plt.contourf(t1, t2, JR, levels=levelsJR, alpha=0.9)
    plt.plot(path_JR[:, 0], path_JR[:, 1], "w-o")
    plt.plot(t1r_min, t2r_min, "rs")
    plt.title(title, fontsize=16)
    plt.axis([t1a, t1b, t2a, t2b])

for subplot in (221, 223):
    plt.subplot(subplot)
    plt.ylabel(r"$\theta_2$", fontsize=20, rotation=0)

for subplot in (223, 224):
    plt.subplot(subplot)
    plt.xlabel(r"$\theta_1$", fontsize=20)

plt.show()
```

- 배경 타원: 정규화되지 않은 MSE비용함수(α=0)
- 등고선 : 각 penalty 값
- 흰색 점 : 비용함수에 대한 Batch Gradient Descent 경로.
- Lasso 는 BGD 경로가 먼저 $θ_2$가 0인 값에 도달 후 global minimum값까지 도달할 때까지 gutter를 roll down한다!!
- Ridge의 정규화된 최소값은 정규화 되지 않은 최소값 보다 $θ=0$에 가깝지만, 가중치는 완전히 제거가 되지 않음.

#### Early Stopping

```python
np.random.seed(42)
m = 100
X = 6 * np.random.rand(m, 1) - 3
y = 2 + X + 0.5 * X**2 + np.random.randn(m, 1)

X_train, X_val, y_train, y_val = train_test_split(X[:50], y[:50].ravel(), test_size=0.5, random_state=10)

poly_scaler = Pipeline([
        ("poly_features", PolynomialFeatures(degree=90, include_bias=False)),
        ("std_scaler", StandardScaler()),
    ])

X_train_poly_scaled = poly_scaler.fit_transform(X_train)
X_val_poly_scaled = poly_scaler.transform(X_val)

sgd_reg = SGDRegressor(n_iter=1,
                       penalty=None,
                       eta0=0.0005,
                       warm_start=True,
                       learning_rate="constant",
                       random_state=42)

n_epochs = 500
train_errors, val_errors = [], []
for epoch in range(n_epochs):
    sgd_reg.fit(X_train_poly_scaled, y_train)
    y_train_predict = sgd_reg.predict(X_train_poly_scaled)
    y_val_predict = sgd_reg.predict(X_val_poly_scaled)
    train_errors.append(mean_squared_error(y_train_predict, y_train))
    val_errors.append(mean_squared_error(y_val_predict, y_val))

best_epoch = np.argmin(val_errors)
best_val_rmse = np.sqrt(val_errors[best_epoch])

plt.annotate('Best model',
             xy=(best_epoch, best_val_rmse),
             xytext=(best_epoch, best_val_rmse + 1),
             ha="center",
             arrowprops=dict(facecolor='black', shrink=0.05),
             fontsize=16,
            )

best_val_rmse -= 0.03  # just to make the graph look better
plt.plot([0, n_epochs], [best_val_rmse, best_val_rmse], "k:", linewidth=2)
plt.plot(np.sqrt(val_errors), "b-", linewidth=3, label="Validation set")
plt.plot(np.sqrt(train_errors), "r--", linewidth=2, label="Training set")
plt.legend(loc="upper right", fontsize=14)
plt.xlabel("Epoch", fontsize=14)
plt.ylabel("RMSE", fontsize=14)
plt.show()
```

- early stopping : 경사하강법과 같은 반복학습 알고리즘은 유효성 에러(validation error)가 최소에 도달하는 순간 학습을 멈춰야 한다는 것.
- validation set의 오류가 늘어나는 시점 : over-fit이 시작된다! > 학습을 멈춰야함

```python
from sklearn.base import clone
sgd_reg = SGDRegressor(n_iter=1, warm_start=True, penalty=None,
                       learning_rate="constant", eta0=0.0005, random_state=42)

minimum_val_error = float("inf")
best_epoch = None
best_model = None
for epoch in range(1000):
    sgd_reg.fit(X_train_poly_scaled, y_train)  # continues where it left off
    y_val_predict = sgd_reg.predict(X_val_poly_scaled)
    val_error = mean_squared_error(y_val_predict, y_val)
    if val_error < minimum_val_error:
        minimum_val_error = val_error
        best_epoch = epoch
        best_model = clone(sgd_reg)
```
