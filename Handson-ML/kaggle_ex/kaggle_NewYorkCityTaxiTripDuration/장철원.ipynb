```python
train = pd.read_csv("C:/Users/cheolwon/Documents/Kaggle/NewYork_taxi/train.csv", header = 0)
test = pd.read_csv("C:/Users/cheolwon/Documents/Kaggle/NewYork_taxi/test.csv", header = 0)

train.fillna(0)
test.fillna(0)
```

```python
# distance between pickup and dropoff place
train['distance']=np.sqrt(( (train['pickup_longitude'] - train['dropoff_longitude'])**2 ) 
                              + ((train['pickup_latitude'] - train['dropoff_latitude'])**2)) 
test['distance']=np.sqrt(( (test['pickup_longitude'] - test['dropoff_longitude'])**2 ) 
                             + ((test['pickup_latitude'] - test['dropoff_latitude'])**2)) 

plt.hist(train['distance'], bins = 100)
plt.xlim((0, 0.5))
plt.grid()
plt.show()
```

```python
# K-means Clustering(training Data)
pickup = pd.concat([train['pickup_longitude'], train['pickup_latitude']], axis=1)
dropoff = pd.concat([train['dropoff_longitude'], train['dropoff_latitude']], axis=1)

k_means = KMeans(init='k-means++', n_clusters=20, n_init=10)
t0 = time.time()
kmeans_fit_11 = k_means.fit(pickup)
kmeans_fit_12 = k_means.fit(dropoff)
t_batch = time.time() - t0
print(t_batch)
```

```python
pred_pick = kmeans_fit_11.predict(pickup)
pred_drop = kmeans_fit_12.predict(dropoff)

train['pick'] = pred_pick
train['drop'] = pred_drop
```

```python
# K-means Clustering(test Data)
pickup_2 = pd.concat([test['pickup_longitude'], test['pickup_latitude']], axis=1)
dropoff_2 = pd.concat([test['dropoff_longitude'], test['dropoff_latitude']], axis=1)

k_means = KMeans(init='k-means++', n_clusters=20, n_init=10)
t0 = time.time()
kmeans_fit_21 = k_means.fit(pickup_2)
kmeans_fit_22 = k_means.fit(dropoff_2)
t_batch = time.time() - t0
print(t_batch)
```

```python
pred_pick_2 = kmeans_fit_21.predict(pickup_2)
pred_drop_2 = kmeans_fit_22.predict(dropoff_2)

test['pick'] = pred_pick_2
test['drop'] = pred_drop_2
```

```python
# training set을 다시 training set 과 test set으로 나누기
train['train'] = np.random.rand(len(train)) < 0.7
train2 = train[train.train == 1]
test2 = train[train.train == 0]
```

```python
x_train = train2.iloc[:,[1,4,12,13,14]]
y_train = train2['trip_duration']

x_test = test2.iloc[:,[1,4,12,13,14]]
y_test = test2['trip_duration']
```

```python
# gradient boosting regressor in sklearn
params = {'n_estimators' : 100, 'max_depth':4, 'min_samples_split':2, 'learning_rate':0.01, 'loss':'ls'}

clf = ensemble.GradientBoostingRegressor(**params)

start = datetime.now()
clf.fit(x_train, y_train)
print("This took", datetime.now() - start)
```

```python
mse = mean_squared_error(y_test, clf.predict(x_test))
print("MSE: %.4f" % mse)
```
